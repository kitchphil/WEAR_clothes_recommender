## Clean and Move personal details to being more usable

```python
import pandas as pd
from datetime import date
import re
```

```python
####################IMPORT LIBRARIES####################
import pandas as pd
from datetime import date
import re


####################GET INPUT FROM USER####################
username = input('username: ')
gender = input('M or F: ')
dob = input('DOB - (DD/MM/YYYY): ')
#age = input('Age:')
height = input('Height(cms): ')
weight = input('Weight(kgs): ')
home = input('Where do you live? (City, Country) ')
sweat = input('Do you sweat easily? Y/N: ')
coldhf = input('Do you often have cold hands or feet? Y/N: ')
#burns = input('Do you burn Easily in the sun? Y/N')
exercise = input ('how often do you exercise? ')


####################TRANSFORM INPUT TO DATAFRAME####################
userinfo = pd.DataFrame({'user' : username
                   ,'gender':[gender]
                   ,'dob':[dob]
                   ,'height_cm':[height]
                   ,'weight_kg':[weight]
                   ,'currentlocation':[home]
                   ,'sweats':[sweat]
                   ,'coldhf':[coldhf]
                   #,'burnsinsun':[burns]
                   ,'exercise':[exercise]}) 



####################CLEAN INPUT DATA####################

#if height cm contains . , or ' <-- indicates feet' inches
for i in userinfo['height_cm']:
    if '.' in i:
            if "cm" in i:
                userinfo['height_cm'] = userinfo['height_cm'].replace('cm',' ',regex = True)
                userinfo['height_cm'] = userinfo['height_cm'].astype(float)
                userinfo['height_cm'] = userinfo['height_cm'] * 100
                userinfo['height_cm'] = userinfo['height_cm'].astype(int)
            else:
                userinfo['height_cm'] = userinfo['height_cm'].astype(float)
                userinfo['height_cm'] = userinfo['height_cm'] * 100
                userinfo['height_cm'] = userinfo['height_cm'].astype(int)
                
    elif ',' in i:
            if "cm" in i:
                userinfo['height_cm'] = userinfo['height_cm'].replace('cm',' ',regex = True)
                userinfo['height_cm'] = userinfo['height_cm'].replace(",",'.', regex = True)
                userinfo['height_cm'] = userinfo['height_cm'].astype(float)
                userinfo['height_cm'] = userinfo['height_cm'] * 100
                userinfo['height_cm'] = userinfo['height_cm'].astype(int)
            else:
                userinfo['height_cm'] = userinfo['height_cm'].replace(",",'.', regex = True)
                userinfo['height_cm'] = userinfo['height_cm'].astype(float)
                userinfo['height_cm'] = userinfo['height_cm'] * 100
                userinfo['height_cm'] = userinfo['height_cm'].astype(int)
    
    #Convert feet and inches
    elif "'" in i:
        userinfo['height_cm'] = round((userinfo['height_cm'].str.split("'", n = 1, expand = True)[0].astype(int) * 30.48)+(userinfo['height_cm'].str.split("'", n = 1, expand = True)[1].astype(int) * 2.54))
        userinfo['height_cm'] = userinfo['height_cm'].astype(int)

        

 ##Calculate age of person##
for i in userinfo['dob']:
    if '/' in i:
        userinfo['dob'] = re.sub('/','',i)
        userinfo['dob'] = (userinfo['dob'].str[0:2]+'-'+userinfo['dob'].str[2:4]+'-'+userinfo['dob'].str[4:9])
        userinfo['dob'] = pd.to_datetime(userinfo['dob'], format = "%d-%m-%Y")
        userinfo['now'] = pd.Timestamp("now")
        userinfo['now'] = userinfo['now'].astype(str).str[0:10]
        userinfo['now'] = pd.to_datetime(userinfo['now'])
        userinfo['now'] = (userinfo['now'] - userinfo['dob'])/365.242199
        userinfo['age'] = userinfo['now'].astype(str).str[0:3]
        userinfo['age'] = userinfo['age'].astype(int)
        userinfo = userinfo.drop(columns='now')
    else:
        userinfo['dob'] = (userinfo['dob'].str[0:2]+'-'+userinfo['dob'].str[2:4]+'-'+userinfo['dob'].str[4:9])
        userinfo['dob'] = pd.to_datetime(userinfo['dob'], format = "%d-%m-%Y")
        userinfo['now'] = pd.Timestamp("now")
        userinfo['now'] = userinfo['now'].astype(str).str[0:10]
        userinfo['now'] = pd.to_datetime(userinfo['now'])
        userinfo['now'] = (userinfo['now'] - userinfo['dob'])/365.242199
        userinfo['age'] = userinfo['now'].astype(str).str[0:3]
        userinfo['age'] = userinfo['age'].astype(int)
        userinfo = userinfo.drop(columns='now')


#####Calculate BMI#####
userinfo['height_cm'] = userinfo['height_cm'].astype(float)
userinfo['weight_kg'] = userinfo['weight_kg'].astype(float)
userinfo['bmi'] = round(userinfo['weight_kg']/((userinfo['height_cm']/100) ** 2),2)


#####Separate Home Column into two Columns#####
userinfo['currentcity'] = userinfo['currentlocation'].str.split(", ", n = 1, expand = True)[0]
userinfo['currentcountry'] = userinfo['currentlocation'].str.split(", ", n = 1, expand = True)[1]


#-----------------WE NOW MOVE ONTO SCORING THE DIFFERENT FACTORS TO DETERMINE THE mytempÂ© SCORE-----------------#



#############SCORE GENDER#############

for i in userinfo['gender']:
    if 'M' in i:
        userinfo['genscore'] = 1
    elif 'F' in i: 
        userinfo['genscore'] = -2

        
#############SCORE AGE#############       

for i in userinfo['gender']:
    if 'M' in i:
        for j in userinfo['age']:
            if j < 18:
                userinfo['agescore'] = 0
            elif j <= 24:
                userinfo['agescore'] = 0
            elif j <= 29:
                userinfo['agescore'] = 0
            elif j <= 34:
                userinfo['agescore'] = 0
            elif j <= 39:
                userinfo['agescore'] = 0  
            elif j <= 44:
                userinfo['agescore'] = 0
            elif j <= 49:
                userinfo['agescore'] = 0    
            elif j <= 54:
                userinfo['agescore'] = 0
            elif j <= 59:
                userinfo['agescore'] = 0 
            elif j <= 64:
                userinfo['agescore'] = -1
            elif j <= 69:
                userinfo['agescore'] = -1     
            elif j <= 74:
                userinfo['agescore'] = -2
            elif j <= 79:
                userinfo['agescore'] = -2               
            elif j >= 80:
                userinfo['agescore'] = -3               
    
    elif 'F' in i:
        for k in userinfo['age']:
            if k < 18:
                userinfo['agescore'] = 0
            elif k <= 24:
                userinfo['agescore'] = 0
            elif k <= 29:
                userinfo['agescore'] = 0
            elif k <= 34:
                userinfo['agescore'] = 0
            elif k <= 39:
                userinfo['agescore'] = 0  
            elif k <= 44:
                userinfo['agescore'] = 0
            elif k <= 49:
                userinfo['agescore'] = 3    
            elif k <= 54:
                userinfo['agescore'] = 3
            elif k <= 59:
                userinfo['agescore'] = 0 
            elif k <= 64:
                userinfo['agescore'] = -1
            elif k <= 69:
                userinfo['agescore'] = -1     
            elif k <= 74:
                userinfo['agescore'] = -2
            elif k <= 79:
                userinfo['agescore'] = -2               
            elif j >= 80:
                userinfo['agescore'] = -3  
        

        
#############SCORE & LABEL BMI#############
for i in userinfo['bmi']:
    if i <= 18:
        userinfo['bmi'] = 'underweight'
        userinfo['bmiscore'] = -1
    elif i > 18:
        userinfo['bmi'] = 'healthy'
        userinfo['bmiscore'] = 0
    elif i > 24:
        userinfo['bmi'] = 'overweight'
        userinfo['bmiscore'] = 3
    elif i > 30:
        userinfo['bmi'] = 'obese'
        userinfo['bmiscore'] = 5
        


#############SCORE CITY#############
#city will always be 0. If the compared city changes, the temperature difference / 2 will be added to the overall score

for i in userinfo['currentlocation']:
    userinfo['cityscore'] = 0
    

    
#############SCORE SWEATS#############

for i in userinfo['sweats']:
    if 'Y' in i:
        userinfo['sweatscore'] = 1
    elif 'N' in i:
        userinfo['sweatscore'] = 0
        
        

#############SCORE COLDHANDS&FEET#############

for i in userinfo['coldhf']:
    if 'Y' in i:
        userinfo['coldhfscore'] = -2
    elif 'N' in i:
        userinfo['coldhfscore'] = 0
        

        
#############SCORE BURNS#############

##for i in userinfo['burnsinsun']:
#    if 'Y' in i:
#        userinfo['burnscore'] = 1
#    elif 'N' in i:
#        userinfo['burnscore'] = 0
        
        
        
#############SCORE EXERCISE#############

for i in userinfo['exercise']:
    if 'Never' in i:
        userinfo['exercisescore'] = 0
    elif '1 or 2 times per month' in i:
        userinfo['exercisescore'] = 0      
    elif 'Once per week' in i:
        userinfo['exercisescore'] = 0  
    elif 'A couple of times per week' in i:
        userinfo['exercisescore'] = 0          
    elif 'Every other day' in i:
        userinfo['exercisescore'] = 1          
    elif 'Every day' in i:
        userinfo['exercisescore'] = 1  
    
    

#############SCORE mytempÂ©#############
    
userinfo['mytemp'] = userinfo['genscore'] + userinfo['agescore'] + userinfo['bmiscore'] + userinfo['cityscore'] + userinfo['sweatscore'] + userinfo['coldhfscore'] + userinfo['exercisescore'] #+ userinfo['burnscore']
    

    #all
#userinfo = userinfo[['user','gender','age','bmi','currentcity','sweats','coldhf','burnsinsun','exercise','genscore', 'agescore','bmiscore', 'cityscore', 'sweatscore', 'coldhfscore', 'burnscore', 'exercisescore', 'mytemp']]

#w/o burn
userinfo = userinfo[['user','gender','age','bmi','currentlocation','currentcity','currentcountry','sweats','coldhf','exercise','genscore', 'agescore','bmiscore', 'cityscore', 'sweatscore', 'coldhfscore', 'exercisescore', 'mytemp']]
        
```
